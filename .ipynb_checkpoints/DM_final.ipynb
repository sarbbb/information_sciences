```python
from mlxtend.plotting import plot_decision_regions
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import pickle
```

```python
# 載入資料
diabetes_data = pd.read_csv("diabetes.csv")
diabetes_data.head()
```

```python
# verbose的參數在這裡可不寫，verbose=True顯示所有Column的訊息
diabetes_data.info(verbose=True)
```

```python
# 用isnull()搭配sum()可查看是否有null值 (這段可以省略，因為上面df.info()的資料裡已經有顯示欄位裡是否有null值)
diabetes_data.isnull().sum().sort_values(ascending=False)
```

```python
# describe()的count可以查看該欄位是否有null值，因為有null值count不計算
# describe()僅可以查看『數值型』資料，如果欄位是『類別型』資料則不顯示
# 如果連『類別型』資料都要顯示，則要使用參數 include="all"
# 語法：diabetes_data.describe(include="all")
diabetes_data.describe()    # 查看資料時，發現欄位['Glucose','BloodPressure','SkinThickness','Insulin','BIM']的『最小值』為0
```

```python
# 查看欄位裡0值的情況，發現欄位['Glucose','BloodPressure','SkinThickness','Insulin','BIM']這幾個欄位裡不應該有資料是0
# 表示，這些欄位裡的0應該是『NaN』
(diabetes_data == 0).astype(int).sum(axis=0)
```

```python
# 建立一個新的diabetes_data_copy，使新的dataset的操作不會影響到原本的dataset
diabetes_data_copy = diabetes_data.copy(deep = True)
# 將新的dataset的欄位['Glucose','BloodPressure','SkinThickness','Insulin','BMI']的欄位值是0的填上NaN (用replace(0,np.NaN))
diabetes_data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)

## 計算有多少NaN值(表示缺失值)
print(diabetes_data_copy.isnull().sum())
```

```python
# 查看新的dataset(diabetes_data_copy)，確認0值都被NaN取代 => 可以看到『Insulin』中有許多『NaN』
diabetes_data_copy
```

```python
# 用missingno繪製直方圖，查看缺失值的情況
import missingno as msno
p=msno.bar(diabetes_data_copy)
```

```python
#查看缺失值的比率
((diabetes_data_copy.isnull().sum())/diabetes_data_copy.shape[0]).sort_values(ascending=False).map(lambda x:"{:.2%}".format(x))
```

```python
# 用中位數填補NaN值
diabetes_data_copy['Glucose'].fillna(diabetes_data_copy['Glucose'].median(), inplace = True)
diabetes_data_copy['BloodPressure'].fillna(diabetes_data_copy['BloodPressure'].median(), inplace = True)
diabetes_data_copy['SkinThickness'].fillna(diabetes_data_copy['SkinThickness'].median(), inplace = True)
diabetes_data_copy['Insulin'].fillna(diabetes_data_copy['Insulin'].median(), inplace = True)
diabetes_data_copy['BMI'].fillna(diabetes_data_copy['BMI'].median(), inplace = True)
```

```python
# 查看填補平均值後的dataset
diabetes_data_copy
```

```python
# 查看填補null後的直方圖
p = diabetes_data_copy.hist(figsize = (20,20))
```

```python
## null count analysis
# 繪製直條圖，查看缺失值
p=msno.bar(diabetes_data)
```

```python
color_wheel = {1:"#C0C0C0",2:"#73B839"}
colors = diabetes_data["Outcome"].map(lambda x: color_wheel.get(x+1))
# 查看 Outcome的值
print(diabetes_data.Outcome.value_counts())
p=diabetes_data.Outcome.value_counts().plot(kind="bar",color=colors) # 沒病的case > 有病的 很多(快要2倍)
```

```python
# 查看兩兩特徵之間的關係
p=sns.pairplot(diabetes_data_copy, hue = 'Outcome')
```

```python
# 查看熱力圖
plt.figure(figsize=(12,10))  # on this line I just set the size of figure to 12 by 10.
p=sns.heatmap(diabetes_data_copy.corr(), annot=True,cmap ='RdYlGn')  # seaborn has very simple solution for heatmap
```

```python
X =  pd.DataFrame(diabetes_data_copy.drop(["Outcome"],axis = 1),
        columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
X.head()
```

```python
#X = diabetes_data.drop("Outcome",axis = 1)
y = diabetes_data_copy.Outcome
```

```python
# 拆分資料為訓練集和測試集
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42, stratify=y)
```

<font face="黑体" color=blue bold=true size=7>**模型訓練 & 模型評估**</font>

```python
# 準備模型訓練和評估
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
models = {}
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = np.arange(1,15)

train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    
    knn = KNeighborsClassifier(n_neighbors=k)

    knn.fit(X_train, y_train)
    
    train_accuracy[i] = knn.score(X_train, y_train)
    
    test_accuracy[i] = knn.score(X_test, y_test) 
```

```python
#畫圖，選取合適的k值
plt.figure(figsize=(12,5))
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy',marker='*')
plt.plot(neighbors, train_accuracy, label='Training accuracy',marker='o')
plt.legend()  #繪製圖例
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()  # k=8測試集Accuracy最高
```

```python
# 將k值設定為6
models['KNN']  = KNeighborsClassifier(6)
models['KNN'].fit(X_train,y_train)
```

```python
# 繪製混淆矩陣
y_pred = models['KNN'].predict(X_test)
confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
# 評估KNN模型
print(classification_report(y_test,y_pred))
```

```python
# 模型準確度
print("Accuracy:",accuracy_score(y_test, y_pred))
```

```python
# 繪製ROC曲線
y_pred_proba = models['KNN'].predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='Knn')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('Knn(n_neighbors=8) ROC curve')
plt.show()
```

```python
# 計算ROC曲線下面積
roc_auc_score(y_test,y_pred_proba)
```

<font face="黑体" color=blue bold=true size=7>**決策應用 & 模型部署**</font>

```python
list_accuracy=[]
for name, model in models.items():
        list_accuracy.append(models[name].score(X_test,y_test))
```

```python
list_ROC=[]
for name, model in models.items():
    y_pred_proba = models[name].predict_proba(X_test)[:,1]
    list_ROC.append(roc_auc_score(y_test,y_pred_proba))
```

```python
Accuracy = pd.Series(list_accuracy, index=[name for name in models])
ROC_Curve_area = pd.Series(list_ROC, index=[name for name in models])
pd.concat([Accuracy, ROC_Curve_area], axis=1,keys= ['Accuracy', 'ROC Curve area']).sort_values(by='Accuracy',ascending=False)
```

```python
plt.figure(0).clf()
for name, model in models.items():
    y_pred_proba = models[name].predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    auc=round(roc_auc_score(y_test,y_pred_proba),3)
    
    plt.plot(fpr,tpr,label=name+":"+str(auc))

    plt.legend(loc=0)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1-Specificity(False Positive Rate)')
plt.ylabel('Sensitivity(True Positive Rate)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()  
```

```python
plt.figure(figsize=(12,5))
x = np.array([name for name in models])
y1 = np.array(list_accuracy)
y2 = np.array(list_ROC)
plt.title('compare models')
plt.plot(x, y1, label='Accuracy',marker='*')
plt.plot(x, y2, label='ROC Curve area',marker='o')
plt.legend()  #繪製圖例
plt.ylabel('Accuracy')
plt.show() 
```
